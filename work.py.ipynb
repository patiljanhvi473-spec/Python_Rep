```python
n=int(input())
m=int(input())
lcm=n
while lcm%n!=0 or lcm%m!=0:
  lcm+=1
print("LCM=",lcm)
```

```python
n=int(input())
m=int(input())
hcf=1
i=1
while i<=n and i<=m:
  if n%i==0 and m%i==0:
    hcf=i
  i+=1
print("HCF=",hcf)
```

```python
n=int(input())
binary=""
while n>0:
  remainder=n%2
  binary=str(remainder)+binary
  n=n//2
  print(binary,end=" ")
```

```python
start=int(input())
end=int(input())
while start<=end:
    print(start,end=" ")
    start=start+2
```

```python
start=int(input())
end=int(input())
while start<=end:
  if start+2>end:
    print(start,end=" ")
  else:
    print(start,end="+")
  start=start+2
```

```python
name = input()
for i in range(len(name)):
    for j in range(len(name)):
        if i == j:
            print(name[i], end=" ")
        else:
            print("*", end=" ")
    print()
```

```python
s=input()
n=len(s)
for i in range(n):
    for j in range(n):
        if i==j:
            print(s[i], end=" ")
        elif i+j==n-1:
            print(s[i], end=" ")
        else:
            print("*", end=" ")
    print()
```
