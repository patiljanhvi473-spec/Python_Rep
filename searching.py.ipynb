```python
a=[1,2,3,4,5,6,7,8,9]
search=int(input())
start=0
end=len(a)-1
mid=(start+end)//2
for i in range(0,len(a)-1):
  if a[mid]==search:
   print(f"Found at {mid} index")
   break
  elif a[mid]>search:
    end=mid-1
    mid=(start+end)//2
  else:
    start=mid+1
    mid=(start+end)//2
```
